## Treinamento

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score, f1_score, roc_auc_score,
    classification_report, confusion_matrix
)
import joblib
```

```python
df = dados2.copy()

target_col = "HeartDisease"
X = df.drop(columns=[target_col])
y = df[target_col]

print("Shape X:", X.shape, "| Shape y:", y.shape)
y.value_counts(normalize=True).rename("freq").to_frame()
```

```python
cat_cols = ["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]
num_cols = [c for c in X.columns if c not in cat_cols]

print("Categ√≥ricas:", cat_cols)
print("Num√©ricas:", num_cols)
```

```python
preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), cat_cols),
    ]
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
print("Train:", X_train.shape, "| Test:", X_test.shape)
```

```python
models = {
    "LogReg": Pipeline([
        ("prep", preprocess),
        ("clf", LogisticRegression(max_iter=1000))  # teste class_weight="balanced"
    ]),
    "RandomForest": Pipeline([
        ("prep", preprocess),
        ("clf", RandomForestClassifier(n_estimators=400, random_state=42))
    ]),
    "GradientBoosting": Pipeline([
        ("prep", preprocess),
        ("clf", GradientBoostingClassifier(random_state=42))
    ]),
    "SVM-RBF": Pipeline([
        ("prep", preprocess),
        ("clf", SVC(kernel="rbf", probability=True, C=1.0, gamma="scale", random_state=42))
    ]),
    "KNN": Pipeline([
        ("prep", preprocess),
        ("clf", KNeighborsClassifier(n_neighbors=7))
    ]),
}

list(models.keys())
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_results = {}
print("CV ROC-AUC (m√©dia ¬± desvio):")
for name, pipe in models.items():
    scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="roc_auc")
    cv_results[name] = {"roc_auc_mean": scores.mean(), "roc_auc_std": scores.std()}
    print(f" - {name}: {scores.mean():.3f} ¬± {scores.std():.3f}")

sorted_cv = sorted(cv_results.items(), key=lambda kv: kv[1]["roc_auc_mean"], reverse=True)
sorted_cv[:3]
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, roc_auc_score

def eval_model(pipe, X_train, y_train, X_test, y_test, threshold=0.5):
    pipe.fit(X_train, y_train)
    proba = pipe.predict_proba(X_test)[:, 1]
    pred = (proba >= threshold).astype(int)

    acc = accuracy_score(y_test, pred)
    f1  = f1_score(y_test, pred)
    auc = roc_auc_score(y_test, proba)

    
    rep = classification_report(y_test, pred, output_dict=True, zero_division=0)
    
    cm = confusion_matrix(y_test, pred)

    return {
        "accuracy": round(acc, 3),
        "f1": round(f1, 3),
        "roc_auc": round(auc, 3),
        "p0": round(rep["0"]["precision"], 3),
        "r0": round(rep["0"]["recall"], 3),
        "p1": round(rep["1"]["precision"], 3),
        "r1": round(rep["1"]["recall"], 3),
        "cm": cm,
        "model": pipe
    }

results = {}
for name, pipe in models.items():
    results[name] = eval_model(pipe, X_train, y_train, X_test, y_test, threshold=0.5)

def print_leaderboard(res):
    header = f"{'Modelo':<16} {'ACC':>6} {'F1':>6} {'AUC':>6} | {'P0':>6} {'R0':>6} | {'P1':>6} {'R1':>6}"
    print(header)
    print("-"*len(header))
    # ordena por AUC (desc)
    for name, r in sorted(res.items(), key=lambda kv: kv[1]["roc_auc"], reverse=True):
        print(f"{name:<16} {r['accuracy']:>6.3f} {r['f1']:>6.3f} {r['roc_auc']:>6.3f} | {r['p0']:>6.3f} {r['r0']:>6.3f} | {r['p1']:>6.3f} {r['r1']:>6.3f}")

print_leaderboard(results)
```

```python
best_name, best_res = max(results.items(), key=lambda kv: kv[1]["roc_auc"])

print(f"Melhor por ROC-AUC: {best_name}")
print(f"ACC={best_res['accuracy']:.3f} | F1={best_res['f1']:.3f} | AUC={best_res['roc_auc']:.3f}")
print(f"P0={best_res['p0']:.3f} | R0={best_res['r0']:.3f} | P1={best_res['p1']:.3f} | R1={best_res['r1']:.3f}")
print("Matriz de confus√£o [ [TN FP] ; [FN TP] ]:")
print(best_res["cm"])
```

# üìå Conclus√£o da Compara√ß√£o de Modelos

A avalia√ß√£o dos cinco modelos treinados ‚Äî **Logistic Regression**, **Random Forest**, **Gradient Boosting**, **SVM-RBF** e **KNN** ‚Äî mostra que todos alcan√ßaram desempenho elevado, mas com diferen√ßas importantes entre eles.

### üîπ **1. Melhor desempenho geral**
O modelo **SVM-RBF** apresentou o **maior ROC-AUC (0.943)**, indicando a melhor capacidade de separar corretamente pacientes com e sem doen√ßa card√≠aca.  
Ele tamb√©m manteve **F1 = 0.900**, demonstrando equil√≠brio entre precis√£o e recall.

### üîπ **2. Precis√£o e Sensibilidade**
- Para a classe **0** (sem doen√ßa), a SVM obteve:
  - **P0 = 0.896**
  - **R0 = 0.841**

  Isso mostra que o modelo acerta muitos negativos, mas ainda comete alguns falsos positivos.

- Para a classe **1** (com doen√ßa, a classe mais importante clinicamente), a SVM manteve:
  - **P1 = 0.879**
  - **R1 = 0.922**

  Ou seja, o modelo √© **muito bom em identificar corretamente quem realmente tem a doen√ßa**, reduzindo o risco de falsos negativos.

### üîπ **3. Compara√ß√£o com os demais modelos**
Apesar de Random Forest e KNN apresentarem **accuracy e F1 altos (‚âà0.897)**, ambos ficaram atr√°s da SVM no crit√©rio mais relevante para diagn√≥stico ‚Äî **ROC-AUC**, que mede qualidade da separa√ß√£o entre classes.
